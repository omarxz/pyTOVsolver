```python
import re
import numpy as np
```

```python
#Define the regex pattern to match the lines containing a_c and rho_c
pattern = r"Found a_c = ([-+]?\d*\.?\d+) for rho_c = ([-+]?\d*\.?\d+e[+]\d*)"

# Initialize empty lists to store a_c and rho_c values
a_c_values = []
rho_c_values = []

# Read the log file
with open("/Users/omar/Research/Axions/pyTOVsolver/output/EW_axions/extract_failed_run/output_EW_axions.log", "r") as file:
    # Iterate through each line in the file
    for line in file:
        # Use regex to search for the pattern in the line
        match = re.search(pattern, line)
        # If a match is found, extract the values of a_c and rho_c
        if match:
            a_c = float(match.group(1))
            rho_c = float(match.group(2))
            # Append the values to the respective lists
            a_c_values.append(a_c)
            rho_c_values.append(rho_c)

# Convert lists to numpy arrays
a_c_array = np.array(a_c_values)
rho_c_array = np.array(rho_c_values)

# Combine rho_c and a_c arrays into a 2D numpy array
data_array = np.column_stack((rho_c_array, a_c_array))

indices = np.argsort(data_array[:, 0])
data_array = data_array[indices]

data_array
```
